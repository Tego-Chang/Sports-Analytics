```python
def addEliminatedPlayerRecord(df):
    SEASON_MAX = 2019
    # check all players' names, identify their last season
    for names in df['PLAYER'].unique():
        if (df.loc[df.PLAYER == names].SEASON.max() < SEASON_MAX):
            season_eliminated = (df.loc[df.PLAYER == names].SEASON.max()) + 1
            for ss in range(int(season_eliminated), (SEASON_MAX + 1)):
                # add records to the season of SEASON_MAX with all other fields as 0, 
                # except the age plus 1 per season
                df = df.append(pd.Series(), ignore_index=True)
                
                age = df.loc[df.PLAYER == names].AGE.max() + 1
                df.iloc[-1, 0] = names
                df.iloc[-1, 2] = ss
                df.iloc[-1, 3] = "ELM"
                df.iloc[-1, 4] = age
                df.iloc[-1, -1] = df_s_cap[df_s_cap['SEASON']==ss].SALARY_CAP
                df = df.fillna(0)
    return df
```
